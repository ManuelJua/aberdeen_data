```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def get_data_wages(sheet_name,header=6,index_col=0,
            file_name='rtisadec2023.xlsx'):
    df=pd.read_excel(file_name,sheet_name=sheet_name,header=header,
                    index_col=index_col)
    return df
r4_df=get_data_wages("20. Median pay (LA)",header=7)
r3_df=get_data_wages("16. Median pay (NUTS3)")
r2_df=get_data_wages("12. Median pay (NUTS2)")
r1_df=get_data_wages("8. Median pay (NUTS1)")
industry_df=get_data_wages('37. Median pay (NUTS1Sector)',file_name='wages_2024.xlsx')
datazones=pd.read_csv('../data_zones.csv')
```
