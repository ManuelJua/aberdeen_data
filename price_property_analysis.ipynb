```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('residential-properties-sales-and-price.csv')
data_zones=pd.read_csv('data_zones.csv')
```

```python
# df.loc[['Aberdeen City','City of Edinburgh'],:]
pivot=df.pivot_table(columns='Reference Area').rename(columns={'Reference Area':'Year'})
pivot.index.rename('year',inplace=True)
```

```python
aberdeen=data_zones[data_zones['LA_Name']=='Aberdeen City']['IZ2011_Name'].unique()
last_year=pivot[aberdeen].iloc[-1,:]
last_year.sort_values().plot(kind='barh',figsize=(10,10))
plt.grid()
```

```python
cheap_aberdeen=pivot[aberdeen].iloc[-1,:].sort_values(ascending=False).tail(5).index
expensive_aberdeen=pivot[aberdeen].iloc[-1,:].sort_values(ascending=False).head(5).index
pivot[[*expensive_aberdeen,*cheap_aberdeen]].plot(legend=True)
```
